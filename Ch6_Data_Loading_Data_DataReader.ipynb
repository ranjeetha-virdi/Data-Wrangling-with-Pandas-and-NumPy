```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import matplotlib.pyplot as plt
```

```python
start_date = "2020-01-1"
end_date = "2022-08-01"

data = web.DataReader(name="TSLA", data_source='yahoo', start=start_date, end=end_date)
print(data)
data.columns
```

```python
close = data['Close']
ax = close.plot(title='Tesla')
ax.set_xlabel('Date')
ax.set_ylabel('Close')
ax.grid()
plt.show()
```
