```python
x=np.linspace(0,5,11)
y=x**2
```

#there are two methods to plot in matplotlib, functional method and object oriented method

#functional method

```python
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("title")
plt.show()       #print output
```

```python
%matplotlib inline

plt.subplot(1,2,1)
plt.plot(x,y,"r")
plt.subplot(1,2,2)
plt.plot(y,x,"b*-")
```

```python
#object oriented API method

fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig=plt.figure()


axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])       #in the list [0.1,0.1,0.8,0.8] 
                                            #0.1:"left most corner start",
                                            #0.1:"height at which the graph is place from base",
                                             #0.8:"percentage of space occupied by the graph",
                                             #0.8:"height of the graph"
        
        
axes1.plot(x,y)
axes1.set_title("Larger Plot")                                                        
axes2.plot(y,x) 
axes2.set_title("Smaller Plot") 
```

```python
fig=plt.figure()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(8,8))  #figsize in inches (8 inch x 8 inch)
axes[0].plot(x,y,"g*-")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].set_title("XvsY")
axes[1].plot(y,x,"ro-")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Y")
axes[1].set_title("YvsX")
plt.tight_layout()                   #we can iterate through the axes or index call the axes
```

```python
fig=plt.figure(figsize=(5,4))

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
sns.pairplot(iris,palette="magma")
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips,col="time",row="smoker")
g.map(sns.distplot,"total_bill")
```

```python
g = sns.FacetGrid(data=tips,col="smoker", row="time")
g.map(plt.scatter,"total_bill","tip")
```
